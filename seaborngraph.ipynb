```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load sample dataset (Replace with your actual dataset)
data = sns.load_dataset("penguins").dropna()

# Set style
sns.set_style("whitegrid")

# 1. Line Plot
plt.figure(figsize=(6,4))
sns.lineplot(x='flipper_length_mm', y='body_mass_g', data=data, hue='species', marker='o')
plt.title("Line Plot")
plt.show()

# 2. Bar Plot
plt.figure(figsize=(6,4))
sns.barplot(x='species', y='body_mass_g', data=data, ci=None, palette='viridis')
plt.title("Bar Plot")
plt.show()

# 3. Scatter Plot
plt.figure(figsize=(6,4))
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', data=data, hue='species', style='species', palette='deep')
plt.title("Scatter Plot")
plt.show()

# 4. Histogram
plt.figure(figsize=(6,4))
sns.histplot(data['body_mass_g'].dropna(), bins=10, kde=True, color='purple')
plt.title("Histogram")
plt.show()

# 5. Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(x='species', y='body_mass_g', data=data, palette='coolwarm')
plt.title("Box Plot")
plt.show()

# 6. Violin Plot
plt.figure(figsize=(6,4))
sns.violinplot(x='species', y='body_mass_g', data=data, palette='Set2')
plt.title("Violin Plot")
plt.show()

# 7. Heatmap
plt.figure(figsize=(6,4))
corr = data.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1, linecolor='black')
plt.title("Heatmap")
plt.show()

# 8. Pair Plot
sns.pairplot(data, hue='species', palette='husl')
plt.show()

# 9. KDE Plot
plt.figure(figsize=(6,4))
sns.kdeplot(data['body_mass_g'].dropna(), shade=True, color='red')
plt.title("KDE Plot")
plt.show()

# 10. Swarm Plot (Avoiding Overlaps)
plt.figure(figsize=(6,4))
sns.stripplot(x='species', y='body_mass_g', data=data, jitter=True, palette='pastel')
plt.title("Swarm Plot (Alternative)")
plt.show()
```
